```python
import numpy as np
```

```python
# train datas
omega1 = np.array([
    [0.28, 1.31, -6.2],
    [0.07, 0.58, -0.78],
    [1.54, 2.01, -1.63],
    [-0.44, 1.18, -4.32],
    [-0.81, 0.21, 5.73],
    [1.52, 3.16, 2.77],
    [2.20, 2.42, -0.19],
    [0.91, 1.94, 6.21],
    [0.65, 1.93, 4.38],
    [-0.26, 0.82, -0.96]
])

omega2 = np.array([
    [0.011, 1.03, -0.21],
    [1.27, 1.28, 0.08],
    [0.13, 3.12, 0.16],
    [-0.21, 1.23, -0.11],
    [-2.18, 1.39, -0.19],
    [0.34, 1.96, -0.16],
    [-1.38, 0.94, 0.45],
    [-0.12, 0.82, 0.17],
    [-1.44, 2.31, 0.14],
    [0.26, 1.94, 0.08]
])

omega3 = np.array([
    [1.36, 2.17, 0.14],
    [1.41, 1.45, -0.38],
    [1.22, 0.99, 0.69],
    [2.46, 2.19, 1.31],
    [0.68, 0.79, 0.87],
    [2.51, 3.22, 1.35],
    [2.60, 2.44, 0.92],
    [0.64, 0.13, 0.97],
    [0.85, 0.58, 0.99],
    [0.66, 0.51, 0.88]
])

classes = [omega1, omega2, omega3]
class_names = ['ω1', 'ω2', 'ω3']
```

```python
# parzen window estimator
def parzen_density(x, samples, h):
    d = samples.shape[1]    # d = 3
    n = samples.shape[0]    # n = 10
    dis = samples - x       # distance
    square_dis = np.sum(dis**2, axis=1)
    exponent = np.exp(-square_dis / (2 * h**2))
    norm_const = 1.0 / (n * (h**d))
    p_hat = norm_const * np.sum(exponent)
    return p_hat
```

```python
# classifier using bayesian rule
# I assumed that the prior probabilities are equal
def classify_parzen(x, classes, h):
    densities = [parzen_density(x, cls, h) for cls in classes]
    predicted_class = np.argmax(densities)
    return predicted_class, densities
```

## a) h = 1

```python
test_points = np.array([
    [0.50, 1.0, 0.0],       # point 1
    [0.31, 1.51, -0.50],    # point 2
    [-0.3, 0.44, -0.1]      # point 3
])

h = 1.0

for i, x in enumerate(test_points, 1):
    pred, dens = classify_parzen(x, classes, h)
    print(f"Test point {i}: {x}")
    for cname, d in zip(class_names, dens):
        print(f"  {cname}: density = {d:.6e}")
    print(f"Predicted class: {class_names[pred]}\n")
```
