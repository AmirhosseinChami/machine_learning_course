## 1-2) plot E[p_n (x)] versus x

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Expected Parzen estimator
def E_pn(x, a, h):
    x = np.asarray(x)
    out = np.zeros_like(x, dtype=float)

    # 0 <= x <= a
    mask_mid = (x >= 0) & (x <= a)
    out[mask_mid] = (1.0/a) * (1.0 - np.exp(-x[mask_mid] / h))

    # x > a
    mask_high = x > a
    out[mask_high] = (1.0/a) * (np.exp(a / h) - 1.0) * np.exp(-x[mask_high] / h)

    return out

a = 1.0
h_values = [1.0, 0.25, 0.0625]

x = np.linspace(-0.1, 2.0, 1000)

plt.figure()

for h in h_values:
    plt.plot(x, E_pn(x, a, h), label=f"h = {h}")

plt.axvline(0, linestyle='--', linewidth=0.7)
plt.axvline(a, linestyle='--', linewidth=0.7)
plt.title("Expected Parzen estimator for a=1")
plt.xlabel("x")
plt.ylabel("E[ p_n(x) ]")
plt.legend()
plt.grid(True)
```

## 1-4) plot E[p_n (x)] in the range 0<= x <= 0.05

```python
a = 1.0
h = 0.0021714724  # from condition in part 1-3
x = np.linspace(0, 0.05, 1000)

# Expected Parzen estimator
def E_pn(x, a, h):
    xp = np.array(x)
    out = np.zeros_like(xp)
    
    # 0  x  a
    mask = (xp >= 0) & (xp <= a)
    out[mask] = (1/a) * (1 - np.exp(-xp[mask]/h))
    
    # x  a
    mask2 = xp >= a
    out[mask2] = (1/a) * (np.exp(a/h) - 1) * np.exp(-xp[mask2]/h)
    return out

y = E_pn(x, a, h)

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("E[p_n(x)]")
plt.title("Expected Parzen Estimate on [0, 0.05]")
plt.grid(True) 
plt.show()
```
